```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# --- 1. LE MODÈLE MULTI-ÉCHELLES  ---



def simulate_chiarella(NIT, kappa, kappa_3, beta_total, gamma, lst_alpha, 
                       sigmaN, sigmaV, P0, V0, g=0.0011):
    """
    Simule le modèle de Chiarella. 
    Si len(lst_alpha) == 1, se comporte comme le modèle classique.
    """
    P = np.zeros(NIT)
    V = np.zeros(NIT)

    alphas = np.array(lst_alpha)
    K = len(alphas)
    # M stocke le momentum pour chaque échelle i au temps t
    M = np.zeros((K, NIT))
    ##############
    # paramètres apprentissage
    rho = 0.01       # mémoire (0.01–0.1)
    lam = 5      # intensité sélection (1–10)
    # performance cumulée
    Pi = np.zeros(K)
    # beta dynamique initial
    beta = np.ones(K) * beta_total / K

    ###########

    # Bruits
    epsilons = np.random.normal(0, sigmaN, NIT)
    etas = np.random.normal(0, sigmaV, NIT)
    
    # Initialisation
    P[0] = P0; P[1] = P0
    V[0] = V0; V[1] = V0 + g
    
    for t in range(1, NIT - 1):
        chartist_demand = 0
        signals = np.zeros(K)
        for i in range(K):
            signals[i] = np.tanh(gamma * M[i, t])
            chartist_demand += beta[i] * signals[i]
            
        # 2. Mise à jour du Prix 
        P[t+1] = P[t] + kappa * (V[t] - P[t]) + chartist_demand  + kappa_3 *((V[t] - P[t])**3) + epsilons[t+1]
        # 3. Mise à jour de la Valeur Fondamentale 
        V[t+1] = V[t] + g + etas[t+1]

        #######
        returns = P[t+1] - P[t]
        
        for i in range(K):
            perf = signals[i] * returns
            Pi[i] = (1 - rho) * Pi[i] + rho * perf

        expPi = np.exp(lam * Pi - np.max(lam * Pi))  # stabilité numérique
        beta = beta_total * expPi / np.sum(expPi)
        #######
        
        # 4. Mise à jour des signaux de momentum pour t+1
        for i in range(K):
            M[i, t+1] = (1 - alphas[i]) * M[i, t] + alphas[i] * (P[t] - P[t-1])
            
    return P, V

# --- 2. FONCTION D'ANALYSE DES BULLES ---

def get_bubble_stats(P, V):
    """Calcule l'écart (mispricing) et la durée des excursions hors de la valeur fondamentale"""
    delta = P - V
    # On identifie les moments où le prix croise la valeur fondamentale
    signs = np.sign(delta)
    # Suppression des zéros potentiels pour le calcul des changements de signe
    signs[signs == 0] = 1 
    crossings = np.where(np.diff(signs) != 0)[0]
    
    if len(crossings) == 0:
        return np.array([len(P)]), delta
    
    durations = np.diff(np.concatenate([[0], crossings, [len(delta)]]))
    return durations, delta

# --- 3. SIMULATIONS ---

# Paramètres fournis (calibration US)
#params = dict(
#    NIT=3000, 
#    kappa=-0.011,kappa_3 = 0.269, 
#    beta_total=0.018, gamma=36.7,
 #   sigmaN=0.042, sigmaV=0.018,
#    P0=4.41, V0=4.41, g=0.0011
#)
# calib JP
params = dict(
    NIT=2000, 
    kappa=-0.011,kappa_3 = 0.269, 
    beta_total=0.018, gamma=25,
    sigmaN=0.057, sigmaV=0.018,
    P0=7.94, V0=7.94, g=0.0024
)



```

```python
def simuler(N):
    duree_bulle_moy_m = np.zeros(N)
    duree_bulle_moy_s = np.zeros(N)
    for i in range(N):
        alphas_multi = np.linspace(0.05, 0.99, 10)
        P_m, V_m = simulate_chiarella(**params, lst_alpha=alphas_multi)
        P_s, V_s = simulate_chiarella(**params, lst_alpha=[0.7])
        durations_m, delta_m= get_bubble_stats(P_m, V_m)
        durations_s, delta_s= get_bubble_stats(P_s, V_s)
        duree_bulle_moy_m[i] = np.mean(durations_m)
        duree_bulle_moy_s[i] = np.mean(durations_s)
        
    print(f"Durée des Bulles multi (Moyenne: {np.mean(duree_bulle_moy_m):.1f})")
    print(f"Durée des Bulles mono (Moyenne: {np.mean(duree_bulle_moy_s):.1f})")

simuler(100)
```

```python
alphas_multi = np.linspace(0.1, 0.99, 10)
P_m, V_m = simulate_chiarella(**params, lst_alpha=alphas_multi)
P_s, V_s = simulate_chiarella(**params, lst_alpha=[0.7])

fig, axes = plt.subplots(4, 4, figsize=(25, 25))

for idx, (P, V, label) in enumerate([(P_s, V_s, "Mono-alpha (0.7)"), (P_m, V_m, "Multi-alpha")]):
    # --- Calculs préliminaires ---
    # Rendements logarithmiques
    returns = np.diff(np.log(P)) 
    durations, delta= get_bubble_stats(P, V)
    # Volatilité mobile (fenêtre de 50)
    window = 50
    volatility = [np.std(returns[i:i+window]) for i in range(len(returns)-window)]

    # 1. Dynamique Prix (votre code existant)
    axes[2*idx, 0].plot(P, lw=1)
    axes[2*idx, 0].set_title(f"Prix : {label}")

    # 2. Rendements (Le "bruit" du marché)
    axes[2*idx, 1].plot(returns, lw=0.5, color='purple', alpha=0.7)
    axes[2*idx, 1].set_title("Log-Returns (Rendements)")

    # 3. Volatilité (Clusters)
    axes[2*idx, 2].plot(volatility, color='orange')
    axes[2*idx, 2].set_title("Volatilité Mobile (50j)")

    # 4. Distribution des Rendements (L'aspect statistique)
    # On compare avec une loi normale pour voir l'écart
    axes[2*idx, 3].hist(returns, bins=50, density=True, alpha=0.6, color='skyblue')
    axes[2*idx, 3].set_yscale('log') # Log scale pour voir les queues épaisses !
    axes[2*idx, 3].set_title("Dist. Rendements (Log-scale)")

    # 5. Autocorrélation des rendements absolus (Mémoire de la volatilité)
    from statsmodels.tsa.stattools import acf
    abs_acf = acf(np.abs(returns), nlags=40)
    axes[2*idx+1, 0].bar(range(len(abs_acf)), abs_acf, color='teal')
    axes[2*idx+1, 0].set_title("ACF des rendements abs. (Mémoire)")
        # Col 1: Prix vs Valeur
    axes[2*idx+1, 1].plot(P, label="Prix", lw=1, alpha=0.8)
    axes[2*idx+1, 1].plot(V, label="Valeur", linestyle="--", color="black", alpha=0.6)
    axes[2*idx+1, 1].set_title(f"Dynamique : {label}")
    axes[2*idx+1, 1].legend()

    # Col 2: Écarts (Bulles)
    axes[2*idx+1, 2].fill_between(range(len(delta)), delta, 0, where=(delta > 0), color='green', alpha=0.4)
    axes[2*idx+1, 2].fill_between(range(len(delta)), delta, 0, where=(delta < 0), color='red', alpha=0.4)
    axes[2*idx+1, 2].set_title(f"Mispricing (P - V)")

    # Col 3: Distribution des durées
    axes[2*idx+1, 3].hist(durations, bins=20, color='gray', edgecolor='white', alpha=0.7)
    axes[2*idx+1, 3].set_title(f"Durée des Bulles (Moyenne: {np.mean(durations):.1f})")

plt.tight_layout()
```
